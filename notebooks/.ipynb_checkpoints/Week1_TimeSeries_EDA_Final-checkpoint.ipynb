# Project 3: Food & Restaurant Services — Week 1 Time-Series EDA

This notebook focuses on the Week 1 requirements: data ingestion, cleaning, continuous datetime handling, time-series aggregation, trend analysis, seasonality, and autocorrelation.


```python
import os
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.style.use("default")
plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["axes.grid"] = True
```

## 1) Data Ingestion

```python
candidate_paths = [
    "balaji_fast_food_sales_dataset.csv",
    "/mnt/data/balaji_fast_food_sales_dataset.csv"
]

data_path = next((p for p in candidate_paths if os.path.exists(p)), None)
if data_path is None:
    raise FileNotFoundError("Dataset file not found in the expected locations.")

if zipfile.is_zipfile(data_path):
    df = pd.read_excel(data_path)
else:
    try:
        df = pd.read_csv(data_path)
    except UnicodeDecodeError:
        df = pd.read_excel(data_path)

df.head()
```

## 2) Basic Inspection

```python
print("Shape:", df.shape)
display(df.info())
display(df.describe(include="all").T)
```

## 3) Data Cleaning and Time Parsing

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["hour"] = df["hour"].astype(str).str.extract(r"(\d+)")[0].astype(int)

df = df.drop_duplicates().sort_values("date").reset_index(drop=True)

df["day_name"] = df["date"].dt.day_name()
df["month_name"] = df["date"].dt.month_name()

print("Missing values per column:")
print(df.isna().sum())
print("\nDuplicate rows after cleaning:", df.duplicated().sum())
```

## 4) Build a Continuous Daily Time Series

```python
daily_ts = (
    df.groupby(df["date"].dt.floor("D"))
      .agg(
          total_quantity=("quantity", "sum"),
          total_sales=("transaction_amount", "sum"),
          order_count=("order_id", "count")
      )
      .sort_index()
)

full_index = pd.date_range(daily_ts.index.min(), daily_ts.index.max(), freq="D")
daily_ts = daily_ts.reindex(full_index)
daily_ts.index.name = "date"

daily_ts[["total_quantity", "total_sales", "order_count"]] = daily_ts[["total_quantity", "total_sales", "order_count"]].fillna(0)
daily_ts["is_missing_day"] = daily_ts["order_count"].eq(0)

print("Continuous date range:", daily_ts.index.min().date(), "to", daily_ts.index.max().date())
print("Total days in range:", len(daily_ts))
print("Missing/empty days filled with 0:", int(daily_ts["is_missing_day"].sum()))
daily_ts.head()
```

## 5) Overall Demand Trend

```python
fig, ax = plt.subplots()
ax.plot(daily_ts.index, daily_ts["total_quantity"], label="Daily Quantity")
ax.plot(daily_ts.index, daily_ts["total_quantity"].rolling(7).mean(), label="7-day Rolling Mean")
ax.set_title("Overall Daily Demand Trend")
ax.set_xlabel("Date")
ax.set_ylabel("Quantity Sold")
ax.legend()
plt.show()
```

## 6) Weekly Pattern

```python
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

weekly_pattern = daily_ts.copy()
weekly_pattern["day_name"] = weekly_pattern.index.day_name()

weekly_avg = weekly_pattern.groupby("day_name")["total_quantity"].mean().reindex(weekday_order)

fig, ax = plt.subplots()
weekly_avg.plot(kind="bar", ax=ax)
ax.set_title("Average Daily Quantity by Day of Week")
ax.set_xlabel("Day of Week")
ax.set_ylabel("Average Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

## 7) Monthly Pattern

```python
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

monthly_pattern = daily_ts.copy()
monthly_pattern["month_name"] = monthly_pattern.index.month_name()

monthly_total = monthly_pattern.groupby("month_name")["total_quantity"].sum().reindex(month_order)

fig, ax = plt.subplots()
monthly_total.plot(kind="bar", ax=ax)
ax.set_title("Total Quantity Sold by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

## 8) Hourly Demand Pattern

```python
hourly_quantity = df.groupby("hour")["quantity"].sum().sort_index()

fig, ax = plt.subplots()
hourly_quantity.plot(kind="bar", ax=ax)
ax.set_title("Total Quantity Sold by Hour")
ax.set_xlabel("Hour of Day")
ax.set_ylabel("Quantity Sold")
plt.xticks(rotation=0)
plt.show()
```

## 9) Top Selling Items

```python
top_items = df.groupby("item_name")["quantity"].sum().sort_values(ascending=False).head(10)

fig, ax = plt.subplots()
top_items.sort_values().plot(kind="barh", ax=ax)
ax.set_title("Top 10 Items by Quantity Sold")
ax.set_xlabel("Quantity Sold")
ax.set_ylabel("Item Name")
plt.show()
```

## 10) Time-Series Decomposition

```python
ts_for_decomposition = daily_ts["total_quantity"].asfreq("D").fillna(0)

decomp = seasonal_decompose(ts_for_decomposition, model="additive", period=7)

fig = decomp.plot()
fig.set_size_inches(14, 10)
plt.show()
```

## 11) Autocorrelation Analysis

```python
fig, ax = plt.subplots()
plot_acf(ts_for_decomposition, lags=30, ax=ax)
ax.set_title("ACF of Daily Quantity")
plt.show()

fig, ax = plt.subplots()
plot_pacf(ts_for_decomposition, lags=30, ax=ax, method="ywm")
ax.set_title("PACF of Daily Quantity")
plt.show()
```

## 12) Outlier View

```python
fig, ax = plt.subplots()
ax.boxplot(daily_ts["total_quantity"], vert=True)
ax.set_title("Boxplot of Daily Quantity")
ax.set_ylabel("Quantity Sold")
plt.show()
```

## 13) Week 1 Summary Insights

```python
print("Week 1 summary:")
print(f"- Continuous daily timeline created with {len(daily_ts)} days.")
print(f"- Empty/missing days filled with zero: {int(daily_ts['is_missing_day'].sum())}.")
print(f"- Highest average demand day: {weekly_avg.idxmax()} ({weekly_avg.max():.2f}).")
print(f"- Highest total month: {monthly_total.idxmax()} ({monthly_total.max():.2f}).")
print(f"- Peak hour: {int(hourly_quantity.idxmax())}:00 ({hourly_quantity.max():.0f}).")
print("- Top 5 items by quantity:")
display(top_items.head())
```
